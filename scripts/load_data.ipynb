# Pre-process video datasets

```python
pocus_folder = '../covid19_ultrasound/data/pocus_videos/convex/'
pocus_metadata = '../covid19_ultrasound/data/dataset_metadata.csv'

covidus_folder_uncropped = '../COVID-US/data/video/'
covidus_folder_cropped = '../COVID-US/data/video/cropped/' 
# covidus_metadata = '../COVID-US/utils/video_metadata.csv'
covidus_metadata = '../COVID-US/utils/video_metadata_cropped.txt'
```

```python
metadata_df = pd.DataFrame(columns=['source','file_path','class','frames_filename','frame_count'])
metadata_df
```

### POCUS:

```python
# POCUS:

temp_metadata = pd.read_csv(pocus_metadata, header=0)
temp_metadata
```

```python
pocus_files = os.listdir(pocus_folder)
for ind, row in temp_metadata.iterrows():
    # print(row)
    if row['Current location']=='data/pocus_videos/Convex' and row['InThisRepo']=='Yes' and row['Probe']=='convex':
        for file in pocus_files:
            if file.startswith(row['Filename']):
                metadata_df.loc[len(metadata_df.index)] = ['POCUS', pocus_folder+file, row['Label'], 'POCUS'+str(len(metadata_df.index)+1), -1] 
                # print(pocus_folder+file)


```

### COVID-US: don't include for now

```python
# COVID-US:

temp_metadata = pd.read_csv(covidus_metadata, header=0)
covidus_cclicense = ['PocusAtlas', 'Litfl', 'Radiopaedia']
temp_metadata
```

```python
covidus_files_cropped = os.listdir(covidus_folder_cropped)
covidus_files_uncropped = os.listdir(covidus_folder_uncropped)
for ind, row in temp_metadata.iterrows():
    # print(row['id'])
    if row['type']=='lung' and row['probe']=='Convex' and row['source'] in covidus_cclicense:
        found = False
        # print(covidus_folder+row['id']+'_prc.avi')
        for file in covidus_files_cropped:
            if file.startswith(row['id']):
                metadata_df.loc[len(metadata_df.index)] = ['COVID-US',covidus_folder_cropped+file, row['class']] 
                print(covidus_folder_cropped+file)
                found = True
        if not found: # if not found, use uncropped version of video (seems no video from uncropped after filtering)
            for file in covidus_files_uncropped:
                if file.startswith(row['id']):
                    metadata_df.loc[len(metadata_df.index)] = ['COVID-US',covidus_folder_uncropped+file, row['class']] 
                    print("uncropped",covidus_folder_uncropped+file)
                    found = True

```

```python
metadata_df
```

# video to image frames

```python
video_frames_path = '../video_frames/'

for ind, row in metadata_df.iterrows():
  vidcap = cv2.VideoCapture(row['file_path'])
  print(row['file_path'])
  success,image = vidcap.read()
  # plt.imshow(image, interpolation='nearest')
  # plt.show()
  count = 0
  while success:
    image = cv2.resize(image, (224,224)) # 224,224,3
    cv2.imwrite(video_frames_path+row['frames_filename']+"_"+str(count)+".jpg", image)     # save frame as JPEG file
    success,image = vidcap.read()
    count += 1
  metadata_df.at[ind,'frame_count'] = count

  
```

```python
metadata_df.to_csv('metadata_df1.csv',header=True,index=False)
metadata_df
```
